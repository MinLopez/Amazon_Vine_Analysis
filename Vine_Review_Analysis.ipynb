```python
import pandas as pd
import os
```

```python
#File load
data_to_load= "vine_table_202201082338.csv"
vine_df = pd.read_csv(data_to_load, low_memory=False)
```

```python
vine_df.head(5)
```

```python
vine_df.dtypes
```

```python
import numpy as np
#using https://www.py4u.net/discuss/156627
vine_df['total_votes'] = vine_df['total_votes'].fillna(0).astype(np.int64)
vine_df['helpful_votes'] = vine_df['helpful_votes'].fillna(0).astype(np.int64)
vine_df.dtypes
```

```python
Above_20_df = vine_df[vine_df["total_votes"]>=20]
#https://www.geeksforgeeks.org/ways-to-filter-pandas-dataframe-by-column-values/
Above_20_df
```

```python
helpful = Above_20_df['helpful_votes']
total = Above_20_df['total_votes']


Above_20_df['helpful_total']= helpful/total * 100
```

```python
helpful_total = Above_20_df[Above_20_df["helpful_total"]>=50]
helpful_total
```

```python
Vine_yes_df = helpful_total[helpful_total["vine"]>='Y']
Vine_yes_df
```

```python
Vine_no_df = helpful_total[helpful_total["vine"]>='N']
Vine_no_df
```

```python
#Determine the total number of reviews paid
paid_total = Vine_yes_df['review_id'].count()
paid_total
```

```python
#Determine the total number of reviews unpaid
unpaid_total = Vine_no_df['review_id'].count()
unpaid_total
```

```python
#the number of 5-star reviews paid
star_paid = Vine_yes_df[Vine_yes_df['star_rating']>= 5]
star_paid_count = star_paid['review_id'].count()
star_paid_count
```

```python
#the number of 5-star reviews unpaid
star_unpaid = Vine_no_df[Vine_no_df['star_rating']>= 5]
star_unpaid_count =  star_unpaid['review_id'].count()
star_unpaid_count
```

```python
#percentage of 5-star reviews for the two types of review paid
paid_star_percent = star_paid_count / paid_total * 100
paid_star_percent
```

```python
#percentage of 5-star reviews for the two types of review unpaid
unpaid_star_percent = star_unpaid_count / unpaid_total * 100
unpaid_star_percent
```
